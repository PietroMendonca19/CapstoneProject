**Read the CSV File into a Pandas DataFrame**

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
QUERY = """
SELECT Age, WorkExp FROM main
"""
age_work_df = pd.read_sql_query(QUERY, conn)

plt.figure(figsize=(15, 10))
age_work_df.plot(kind='scatter', x='Age', y='WorkExp', color='red')
plt.title('Relationship between Age and Work Experience')
plt.xlabel('Age range')
plt.xticks(rotation='vertical')
plt.ylabel('Work Experience')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
## Write your code here
QUERY = """
SELECT Age, TimeSearching, Frustration FROM main
"""
bubble = pd.read_sql_query(QUERY, conn)

bubble = bubble[bubble["Age"]!= "Prefer not to say"] # Drop respondents that didn't give their age

# Age mapping
age_map = {
    'Under 18 years old': 16,
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 65
}

bubble['AgeMidPoint'] = bubble['Age'].map(age_map)

# Time Search mapping
bubble = bubble[bubble['TimeSearching'].notna()] #Drop missing values

time_map = {
    'Less than 15 minutes a day': 15,
    '15-30 minutes a day': 23,
    '30-60 minutes a day': 45,
    '60-120 minutes a day': 90,
    'Over 120 minutes a day': 120
}
bubble['TimeSearchingMidPoint'] = bubble['TimeSearching'].map(time_map)

# Create a score for the frustration with the amount of problems
bubble = bubble[bubble['Frustration'].notna()] #Drop missing values
bubble['FrustrationScore'] = bubble['Frustration'].apply(
    lambda x: 0 if 'None of these' in x else len(x.split(';'))
)

# Normalize the age with Min-Max
weight = (bubble['AgeMidPoint']-bubble['AgeMidPoint'].min())/(bubble['AgeMidPoint'].max()-bubble['AgeMidPoint'].min())
plt.figure(figsize=(12,10))
bubble.plot(kind='scatter', x='TimeSearchingMidPoint', y='FrustrationScore', s=weight*15, alpha=0.6, color='darkred')
plt.title('Bubble plot Time Searching vs Frustation with the Age as weight')
plt.xlabel("Time Searching")
plt.ylabel('Frustation Score')
plt.show()
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
QUERY = """
SELECT ResponseId, DatabaseWantToWorkWith FROM main
"""
db_df = pd.read_sql_query(QUERY, conn)

db_df = db_df[['ResponseId','DatabaseWantToWorkWith']].dropna()
db_df = db_df.assign(DataBase=db_df['DatabaseWantToWorkWith'].str.split(';'))
db_df = db_df.explode('DataBase')

top_dbs = db_df['DataBase'].value_counts().nlargest(5)
plt.figure(figsize=(10, 8))
plt.pie(top_dbs, labels=top_dbs.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Want to Learn Next Year')
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
QUERY = """
SELECT Age, CompTotal FROM main
"""
line = pd.read_sql_query(QUERY, conn)

line = line[line["Age"]!= "Prefer not to say"]
line = line[line['CompTotal'].notna()]

filtered_line = line[line['Age'].isin(['45-54 years old', '55-64 years old'])]
median_comp_by_age = filtered_line.groupby('Age')['CompTotal'].median()
median_comp_by_age = median_comp_by_age.loc[['45-54 years old', '55-64 years old']]

plt.figure(figsize=(10, 8))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o')
plt.title('Median Compensation (CompTotal) by Age Group (45–64)', fontsize=14)
plt.xlabel('Age Group')
plt.ylabel('Median Compensation')
plt.show()
```
